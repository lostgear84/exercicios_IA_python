```python
def somar():
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 + num2
    print(f"Resultado da soma: {resultado}")

def subtrair():
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 - num2
    print(f"Resultado da subtração: {resultado}")

def multiplicar():
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    resultado = num1 * num2
    print(f"Resultado da multiplicação: {resultado}")

def dividir():
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))
    if num2 != 0:
        resultado = num1 / num2
        print(f"Resultado da divisão: {resultado}")
    else:
        print("Erro: Divisão por zero não é permitida.")

def verificar_par():
    num = int(input("Digite um número: "))
    if num % 2 == 0:
        print(f"O número {num} é par.")
    else:
        print(f"O número {num} é ímpar.")

def calcular_idade():
    ano_nascimento = int(input("Digite o ano de nascimento: "))
    ano_atual = 2025
    idade = ano_atual - ano_nascimento
    print(f"Sua idade é: {idade} anos")

def calcular_idade_media():
    ano1 = int(input("Digite o ano de nascimento da primeira pessoa: "))
    ano2 = int(input("Digite o ano de nascimento da segunda pessoa: "))
    ano3 = int(input("Digite o ano de nascimento da terceira pessoa: "))
    ano_atual = 2025
    idade1 = ano_atual - ano1
    idade2 = ano_atual - ano2
    idade3 = ano_atual - ano3
    media = (idade1 + idade2 + idade3) / 3
    print(f"A idade média é: {media:.2f} anos")

def verificar_intervalo():
    num = int(input("Digite um número: "))
    if 20 <= num <= 100:
        print(f"O número {num} está entre 20 e 100.")
    else:
        print(f"O número {num} não está entre 20 e 100.")

def verificar_aprovacao():
    nota = float(input("Digite a nota do aluno (0 a 10): "))
    frequencia = float(input("Digite a frequência do aluno (0 a 100): "))
    if nota >= 7 and frequencia >= 75:
        print("Aluno aprovado!")
    else:
        print("Aluno reprovado.")

def verificar_senha_forte():
    senha = input("Digite a senha: ")
    if len(senha) >= 10 and any(c in "!@#$%" for c in senha):
        print("A senha é forte.")
    else:
        print("A senha não é forte. Deve ter no mínimo 10 caracteres e incluir !, @, #, $ ou %.")

def exibir_menu():
    print("\nOpções:")
    print("1) Somar dois números")
    print("2) Subtrair dois números")
    print("3) Multiplicar dois números")
    print("4) Dividir dois números")
    print("5) Verificar se um número informado é par")
    print("6) Calcular a idade de um usuário a partir do seu ano de nascimento")
    print("7) Calcular a idade média de 3 pessoas, a partir dos seus anos de nascimento")
    print("8) Descubra se um número está entre 20 e 100")
    print("9) Identifique se o aluno foi aprovado, com base na nota (mínimo 7) e na frequência (mínimo 75)")
    print("10) Verifique se a senha informada é forte (mínimo 10 caracteres, incluindo !, @, #, $ ou %)")
    print("Escolha uma opção:")

def main():
    while True:
        exibir_menu()
        opcao = input()

        if opcao == "1":
            somar()
        elif opcao == "2":
            subtrair()
        elif opcao == "3":
            multiplicar()
        elif opcao == "4":
            dividir()
        elif opcao == "5":
            verificar_par()
        elif opcao == "6":
            calcular_idade()
        elif opcao == "7":
            calcular_idade_media()
        elif opcao == "8":
            verificar_intervalo()
        elif opcao == "9":
            verificar_aprovacao()
        elif opcao == "10":
            verificar_senha_forte()
        else:
            print("Opção inválida")

        continuar = input("\nDeseja continuar? (s/n): ")
        if continuar.lower() != 's':
            break

if __name__ == "__main__":
    main()
```
